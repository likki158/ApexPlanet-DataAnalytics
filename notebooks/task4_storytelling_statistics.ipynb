# Task 4 - Data Storytelling & Statistical Validation

## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading Dataset

```python
orders = pd.read_csv("../cleaned_data/orders_cleaned.csv")
```

## Business Hypothesis

Hypothesis:
Late deliveries have significantly higher delivery times compared to on-time deliveries.

```python
orders.groupby('is_late_delivery')['delivery_time_days'].mean()
```

## Average Delivery Time Comparison

```python
avg_delivery = orders.groupby('is_late_delivery')['delivery_time_days'].mean()

avg_delivery.plot(kind='bar')

plt.title("Average Delivery Time: Late vs On-Time Deliveries")
plt.xlabel("Late Delivery")
plt.ylabel("Average Delivery Time (Days)")

plt.show()
```

## Statistical Validation using T-Test

```python
from scipy.stats import ttest_ind
```

```python
late = orders[orders['is_late_delivery'] == True]['delivery_time_days']

on_time = orders[orders['is_late_delivery'] == False]['delivery_time_days']
```

```python
t_stat, p_value = ttest_ind(late.dropna(), on_time.dropna())

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

## Interpretation of Statistical Results

The statistical test showed that the difference between late deliveries and on-time deliveries was statistically significant. This indicates that late deliveries are associated with higher delivery times and are not caused by random variation.

## Business Storytelling Insights

### Key Insights
- Most customer orders were delivered successfully.
- Average delivery time was approximately 12 days.
- Late deliveries represented a smaller percentage of total deliveries.
- Certain months experienced operational delivery delays.
- Statistical testing confirmed that late deliveries had significantly higher delivery times.

## Conclusion

This analysis combined exploratory data analysis, dashboarding, and statistical testing to evaluate operational efficiency within the e-commerce dataset. The findings can help businesses improve logistics performance, monitor delivery risks, and enhance customer satisfaction.
